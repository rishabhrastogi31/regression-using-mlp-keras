
In this notebook regression model is implemented using keras multi-level perceptron neural network to predict the strength of concrete using given parameters  the dataset is defined below




<strong>The dataset is about the compressive strength of different samples of concrete based on the volumes of the different ingredients that were used to make them. Ingredients include:</strong>

<strong>1. Cement</strong>

<strong>2. Blast Furnace Slag</strong>

<strong>3. Fly Ash</strong>

<strong>4. Water</strong>

<strong>5. Superplasticizer</strong>

<strong>6. Coarse Aggregate</strong>

<strong>7. Fine Aggregate</strong>

```python
import numpy as np
import pandas as pd
```

```python
concrete=pd.read_csv('concrete_data.csv')
```

```python
concrete.head()
```

```python
cols=concrete.columns 
```

```python
cols!= 'Strength'
```

```python
concrete_data=concrete[cols[cols!= 'Strength'] ]
```

```python
concrete_data.head()
```

```python
target=concrete['Strength']
```

```python
n_cols=concrete_data.shape[1]
```

```python
#data normalization
concrete_data_norm=(concrete_data-concrete_data.min())/(concrete_data.max()-concrete_data.min())
```

```python
concrete_data_norm.head()
```

### Build a neural network 

```python
import keras 
from  keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    model =Sequential()
    model.add(Dense(10, activation= 'relu', input_shape=(n_cols,)) )
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return  model
```

### Train and test the network

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(concrete_data, target, test_size=0.30, random_state=42)
```

```python
model=regression_model()
```

```python
model.fit(X_train, y_train, epochs=100)
```

### Evaluate the model

```python
print(model.evaluate(X_test, y_test))
```

```python
y_pred=model.predict(X_test, batch_size=50)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
msqr=mean_squared_error(y_test, y_pred)
```

```python
print(' mean-squared-error: ', msqr)
```
